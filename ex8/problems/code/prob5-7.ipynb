```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f_0(x):
    return x**2 + 1
def f_1(x):
    return (x-2) * (x-4)
```

```python
# problem 5

x = np.linspace(0,6)
plt.plot(x, f_0(x), label='f0')
plt.plot(x, f_1(x), label='f1')
plt.axvspan(2, 4, color='red', alpha=0.5, label='constraint satisfied')
plt.legend()
plt.savefig('prob5_fig.pdf')
```

```python
# problem 6

def L(x, α):
    return f_0(x) + α * f_1(x)
```

```python
x = np.linspace(0,6)
α = np.array([0, 0.5, 1, 1.5, 2, 3, 4, 5, 8])

for p in α:
    plt.plot(x, L(x,p), label=('alpha=%.1f' % p))

plt.legend()
plt.savefig('prob6_fig.pdf')
```
